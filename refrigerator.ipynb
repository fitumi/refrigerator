```python
import random


def main():
    matrix = create_matrix()

    print('Initial:')
    print_matrix(matrix)

    solve(matrix)


def create_matrix():
    size = 4
    matrix = [
        [bool(random.getrandbits(1)) for __ in range(0, size)]
        for __ in range(0, size)
    ]

    return matrix


def print_matrix(matrix):
    representation = '\n'.join([format_row(row) for row in matrix])
    print(representation)


def format_row(row):
    return ' '.join(get_symbol(item) for item in row)


def get_symbol(item):
    return '+' if item else '-'


def solve(matrix):
    while True:
        switches = get_on_switches(matrix)

        if not switches:
            print('Done!')
            break

        for row, col in switches:
            print('Toggle ({}, {})\n'.format(row, col))
            toggle(matrix, row, col)
            print_matrix(matrix)


def get_on_switches(matrix):
    return [
        (row_index, col_index)
        for row_index, row in enumerate(matrix)
        for col_index, item in enumerate(row)
        if item
    ]


def toggle(matrix, row, col):
    for i in range(0, len(matrix)):
        matrix[i][col] = not matrix[i][col]

    for i in range(0, len(matrix[row])):
        matrix[row][i] = not matrix[row][i]

    matrix[row][col] = not matrix[row][col]


if __name__ == '__main__':
    main()
```
